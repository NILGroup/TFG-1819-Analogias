```python
import spacy
import csv
nlp = spacy.load('es_core_news_sm')


data=[]
cont = 1
csvarchivo = open('salida-definitiva.csv',encoding="utf8",errors='ignore')
entrada = csv.DictReader(csvarchivo,delimiter=";")
csvsalida = open('salida1000palabras.csv', 'w', encoding="utf8")
salida = csv.writer(csvsalida, delimiter=";")



salida.writerow(("PALABRA", "TAG"))

for i in entrada:
    data.append(i['PALABRA'])

for i in range(len(data)):
    doc = nlp(data[i])
    
    for token in doc:
        print(cont,token.text, token.pos_)
        salida.writerow((token.text, token.pos_))
    cont +=1
    
csvarchivo.close()
csvsalida.close()
```
